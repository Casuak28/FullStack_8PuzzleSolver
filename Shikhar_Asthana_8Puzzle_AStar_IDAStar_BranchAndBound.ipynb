```python
root = Tk()
root.title("DTU 2K22_AFI_24 Shikhar 8-Puzzle Problem")

global_counter = 1

button_variables = ["b1","b2","b3","b4","b5","b6","b7","b8","b9"]
button_variables2 = ["b11","b12","b13","b14","b15","b16","b17","b18","b19",
                 "b21","b22","b23","b24","b25","b26","b27","b28","b29",
                "b31","b32","b33","b34","b35","b36","b37","b38","b39"]


button_bfs = ["b11","b12","b13","b14","b15","b16","b17","b18","b19"]
button_dfs = ["b21","b22","b23","b24","b25","b26","b27","b28","b29"]
button_iddfs = ["b31","b32","b33","b34","b35","b36","b37","b38","b39"]

#Max nodes possible for 4 branching factor: 65536
iddfs_max_depth = 20

timer_bfs, timer_dfs, timer_iddfs, comparisons_bfs, comparisons_dfs, comparisons_iddfs = 0,0,0,0,0,0
timer_astar = 0
timer_idastar = 0
timer_bb = 0


#GLOBAL VARIABLE to store sequence of visited nodes in BFS and DFS implementations
visited_states_bfs = []
visited_states_dfs = []
visited_states_iddfs = []
visited_states_idastar = []

visited_states_idastar_local = []

visited_states_bb = []
    
        
#Function to update the grid - A* 
def update_grid_astar():
    
    global visited_states_astar
    
    for state in visited_states_astar:
        b1,b2,b3,b4,b5,b6,b7,b8,b9 = state

        #asssigning the numbers to the tiles
        globals()["b11"]["text"] = str(b1)
        globals()["b12"]["text"] = str(b2)
        globals()["b13"]["text"] = str(b3)
        globals()["b14"]["text"] = str(b4)
        globals()["b15"]["text"] = str(b5)
        globals()["b16"]["text"] = str(b6)
        globals()["b17"]["text"] = str(b7)
        globals()["b18"]["text"] = str(b8)
        globals()["b19"]["text"] = str(b9)

        #assigning the colour
        globals()["b11"].config(bg="Silver")
        globals()["b12"].config(bg="Silver")
        globals()["b13"].config(bg="Silver")
        globals()["b14"].config(bg="Silver")
        globals()["b15"].config(bg="Silver")
        globals()["b16"].config(bg="Silver")
        globals()["b17"].config(bg="Silver")
        globals()["b18"].config(bg="Silver")
        globals()["b19"].config(bg="Silver")

        #assinging the red tile
        if b1 == 9:
            globals()["b11"].config(bg="Red")
            globals()["b11"]["text"] = " "
        elif b2 == 9:
            globals()["b12"].config(bg="Red")
            globals()["b12"]["text"] = " "
        elif b3 == 9:
            globals()["b13"].config(bg="Red")
            globals()["b13"]["text"] = " "
        elif b4 == 9:
            globals()["b14"].config(bg="Red")
            globals()["b14"]["text"] = " "
        elif b5 == 9:
            globals()["b15"].config(bg="Red")
            globals()["b15"]["text"] = " "
        elif b6 == 9:
            globals()["b16"].config(bg="Red")
            globals()["b16"]["text"] = " "
        elif b7 == 9:
            globals()["b17"].config(bg="Red")
            globals()["b17"]["text"] = " "
        elif b8 == 9:
            globals()["b18"].config(bg="Red")
            globals()["b18"]["text"] = " "
        elif b9 == 9:
            globals()["b19"].config(bg="Red")
            globals()["b19"]["text"] = " "
            
        #update the UI
        root.update()
        time.sleep(0.01)        
        
        

#Function to update the grid - IDA*
def update_grid_idastar():
    
    global visited_states_idastar
    
    for state in visited_states_idastar:
        b1,b2,b3,b4,b5,b6,b7,b8,b9 = state

        #asssigning the numbers to the tiles
        globals()["b21"]["text"] = str(b1)
        globals()["b22"]["text"] = str(b2)
        globals()["b23"]["text"] = str(b3)
        globals()["b24"]["text"] = str(b4)
        globals()["b25"]["text"] = str(b5)
        globals()["b26"]["text"] = str(b6)
        globals()["b27"]["text"] = str(b7)
        globals()["b28"]["text"] = str(b8)
        globals()["b29"]["text"] = str(b9)

        #assigning the colour
        globals()["b21"].config(bg="Silver")
        globals()["b22"].config(bg="Silver")
        globals()["b23"].config(bg="Silver")
        globals()["b24"].config(bg="Silver")
        globals()["b25"].config(bg="Silver")
        globals()["b26"].config(bg="Silver")
        globals()["b27"].config(bg="Silver")
        globals()["b28"].config(bg="Silver")
        globals()["b29"].config(bg="Silver")

        #assinging the red tile
        if b1 == 9:
            globals()["b21"].config(bg="Red")
            globals()["b21"]["text"] = " "
        elif b2 == 9:
            globals()["b22"].config(bg="Red")
            globals()["b22"]["text"] = " "
        elif b3 == 9:
            globals()["b23"].config(bg="Red")
            globals()["b23"]["text"] = " "
        elif b4 == 9:
            globals()["b24"].config(bg="Red")
            globals()["b24"]["text"] = " "
        elif b5 == 9:
            globals()["b25"].config(bg="Red")
            globals()["b25"]["text"] = " "
        elif b6 == 9:
            globals()["b26"].config(bg="Red")
            globals()["b26"]["text"] = " "
        elif b7 == 9:
            globals()["b27"].config(bg="Red")
            globals()["b27"]["text"] = " "
        elif b8 == 9:
            globals()["b28"].config(bg="Red")
            globals()["b28"]["text"] = " "
        elif b9 == 9:
            globals()["b29"].config(bg="Red")
            globals()["b29"]["text"] = " "
        
        #update the UI
        root.update()
        time.sleep(0.01)             
        
        
#Function to update the grid - Branch and Bound
def update_grid_bb():
    
    global visited_states_bb
    
    for state in visited_states_bb:
        b1,b2,b3,b4,b5,b6,b7,b8,b9 = state

        #asssigning the numbers to the tiles
        globals()["b31"]["text"] = str(b1)
        globals()["b32"]["text"] = str(b2)
        globals()["b33"]["text"] = str(b3)
        globals()["b34"]["text"] = str(b4)
        globals()["b35"]["text"] = str(b5)
        globals()["b36"]["text"] = str(b6)
        globals()["b37"]["text"] = str(b7)
        globals()["b38"]["text"] = str(b8)
        globals()["b39"]["text"] = str(b9)

        #assigning the colour
        globals()["b31"].config(bg="Silver")
        globals()["b32"].config(bg="Silver")
        globals()["b33"].config(bg="Silver")
        globals()["b34"].config(bg="Silver")
        globals()["b35"].config(bg="Silver")
        globals()["b36"].config(bg="Silver")
        globals()["b37"].config(bg="Silver")
        globals()["b38"].config(bg="Silver")
        globals()["b39"].config(bg="Silver")

        #assinging the red tile
        if b1 == 9:
            globals()["b31"].config(bg="Red")
            globals()["b31"]["text"] = " "
        elif b2 == 9:
            globals()["b32"].config(bg="Red")
            globals()["b32"]["text"] = " "
        elif b3 == 9:
            globals()["b33"].config(bg="Red")
            globals()["b33"]["text"] = " "
        elif b4 == 9:
            globals()["b34"].config(bg="Red")
            globals()["b34"]["text"] = " "
        elif b5 == 9:
            globals()["b35"].config(bg="Red")
            globals()["b35"]["text"] = " "
        elif b6 == 9:
            globals()["b36"].config(bg="Red")
            globals()["b36"]["text"] = " "
        elif b7 == 9:
            globals()["b37"].config(bg="Red")
            globals()["b37"]["text"] = " "
        elif b8 == 9:
            globals()["b38"].config(bg="Red")
            globals()["b38"]["text"] = " "
        elif b9 == 9:
            globals()["b39"].config(bg="Red")
            globals()["b39"]["text"] = " "
        
        #update the UI
        root.update()
        time.sleep(0.01)        
        
        
        
#Function to assign the numbers to each tiles
def button_click(b):
    global global_counter
    
    #check if the tile is empty
    if b["text"] == " " and global_counter < 9:
        b["text"] = global_counter
        global_counter += 1
    elif b["text"] == " " and global_counter == 9:
        global_counter += 1
        b.config(bg="Red")
        
    else:
        messagebox.showerror("8Puzzle","That tile has already been assigned a value. Please choose another tile")

#Function to check whether all tiles have actually been assigned a value or not
def check_input_tiles():
    global global_counter
    
    if global_counter == 10:
        return True
    else:
        messagebox.showerror("8PuzzleEmpty","All tiles must have a numeric value. Please assign proper values to tiles and try again.")
        return False
        
        
#Function to convert user input to start_state
def convert_user_input_to_start_state(b1,b2,b3,b4,b5,b6,b7,b8,b9):
    
    if b1 == " ":
        return [9,b2,b3,b4,b5,b6,b7,b8,b9]
    elif b2 == " ":
        return [b1,9,b3,b4,b5,b6,b7,b8,b9]
    elif b3 == " ":
        return [b1,b2,9,b4,b5,b6,b7,b8,b9]
    elif b4 == " ":
        return [b1,b2,b3,9,b5,b6,b7,b8,b9]
    elif b5 == " ":
        return [b1,b2,b3,b4,9,b6,b7,b8,b9]
    elif b6 == " ":
        return [b1,b2,b3,b4,b5,9,b7,b8,b9]
    elif b7 == " ":
        return [b1,b2,b3,b4,b5,b6,9,b8,b9]
    elif b8 == " ":
        return [b1,b2,b3,b4,b5,b6,b7,9,b9]
    elif b9 == " ":
        return [b1,b2,b3,b4,b5,b6,b7,b8,9]


#Function to find the next move based on the position of red tile
def find_next_moves(state):
    b1,b2,b3,b4,b5,b6,b7,b8,b9 = state
    next_moves = []
    
    if b1 == 9:
        #red tile can move right
        next_moves.append([b2,b1,b3,b4,b5,b6,b7,b8,b9])
        #red tile can move down
        next_moves.append([b4,b2,b3,b1,b5,b6,b7,b8,b9])
    elif b2 == 9:
        #red tile can move left
        next_moves.append([b2,b1,b3,b4,b5,b6,b7,b8,b9])
        #red tile can move right
        next_moves.append([b1,b3,b2,b4,b5,b6,b7,b8,b9])
        #red tile can move down
        next_moves.append([b1,b5,b3,b4,b2,b6,b7,b8,b9])
    elif b3 == 9:
        #red tile can move left
        next_moves.append([b1,b3,b2,b4,b5,b6,b7,b8,b9])
        #red tile can move down
        next_moves.append([b1,b2,b6,b4,b5,b3,b7,b8,b9])
    elif b4 == 9:
        #red tile can move up
        next_moves.append([b4,b2,b3,b1,b5,b6,b7,b8,b9])
        #red tile can move right
        next_moves.append([b1,b2,b3,b5,b4,b6,b7,b8,b9])
        #red tile can move down
        next_moves.append([b1,b2,b3,b7,b5,b6,b4,b8,b9])
    elif b5 == 9:
        #red tile can move up
        next_moves.append([b1,b5,b3,b4,b2,b6,b7,b8,b9])
        #red tile can move left
        next_moves.append([b1,b2,b3,b5,b4,b6,b7,b8,b9])
        #red tile can move right
        next_moves.append([b1,b2,b3,b4,b6,b5,b7,b8,b9])
        #red tile can move down
        next_moves.append([b1,b2,b3,b4,b8,b6,b7,b5,b9])
    elif b6 == 9:
        #red tile can move up
        next_moves.append([b1,b2,b6,b4,b5,b3,b7,b8,b9])
        #red tile can move left
        next_moves.append([b1,b2,b3,b4,b6,b5,b7,b8,b9])
        #red tile moving down
        next_moves.append([b1,b2,b3,b4,b5,b9,b7,b8,b6])
    elif b7 == 9:
        #red tile can move up
        next_moves.append([b1,b2,b3,b7,b5,b6,b4,b8,b9])
        #red tile can move right
        next_moves.append([b1,b2,b3,b4,b5,b6,b8,b7,b9])
    elif b8 == 9:
        #red tile can move up
        next_moves.append([b1,b2,b3,b4,b8,b6,b7,b5,b9])
        #red tile can move left
        next_moves.append([b1,b2,b3,b4,b5,b6,b8,b7,b9])
        #red tile can move right
        next_moves.append([b1,b2,b3,b4,b5,b6,b7,b9,b8])
    elif b9 == 9:
        #red tile can move up
        next_moves.append([b1,b2,b3,b4,b5,b9,b7,b8,b6])
        #red tile can move left
        next_moves.append([b1,b2,b3,b4,b5,b6,b7,b9,b8])
        
    return next_moves





#Function to do BFS traversal
def solve_using_bfs(start_state, goal_state):
    global visited_states_bfs, timer_bfs, comparisons_bfs
    visited_states_bfs = []
    timer_bfs = time.time()
    comparisons_bfs = 0
    q = Queue()
    
    q.put(start_state)
    
    while not q.empty():
        state = q.get()
        if state == goal_state:
            comparisons_bfs += 1
            visited_states_bfs.append(state)
            timer_bfs = (time.time() - timer_bfs) * 10**3
            print("BFS - Goal State Reached")
            #Updating the output grid
            update_grid_bfs()
            return True
        elif state not in visited_states_bfs:
            comparisons_bfs += 1
            visited_states_bfs.append(state)
            neighbouring_states = find_next_moves(state)
            for item in neighbouring_states:
                q.put(item)
    timer_bfs = (time.time() - timer_bfs) * 10**3           
    return False


#Function to do DFS traversal
def solve_using_dfs(start_state,goal_state):
    global visited_states_dfs, timer_dfs, comparisons_dfs
    visited_states_dfs = []
    timer_dfs = time.time()
    comparisons_dfs = 0
    stackk = LifoQueue()
    
    stackk.put(start_state)
    
    while not stackk.empty():
        state = stackk.get()
        if state == goal_state:
            comparisons_dfs += 1
            visited_states_dfs.append(state)
            timer_dfs = (time.time() - timer_dfs) * 10**3  
            print("DFS - Goal State Reached")
            #Updating the output grid
            update_grid_dfs()
            return True
        elif state not in visited_states_dfs:
            comparisons_dfs += 1
            visited_states_dfs.append(state)
            neighbouring_states = find_next_moves(state)
            for item in neighbouring_states:
                stackk.put(item)
    timer_dfs = (time.time() - timer_dfs) * 10**3            
    return False



#Function to do IDDFS traversal(Recursive Implementation)
#Base Function for depth limited depth first search
def depth_limited_dfs(current_state,goal_state, max_depth):
    global visited_states_iddfs, comparisons_iddfs, timer_iddfs
    visited_states_iddfs.append(current_state)
    comparisons_iddfs += 2
    if current_state == goal_state:
        timer_iddfs = (time.time() - timer_iddfs) * 10**3 
        print("IDDFS Goal State reached.")
        #Updating output grid
        update_grid_iddfs()
        return True
    if max_depth<=0:
        return False
    
    neighbouring_states = find_next_moves(current_state)
    for state in neighbouring_states:
        if depth_limited_dfs(state,goal_state,max_depth-1):
            return True
        
    return False

#Function to drive the depth limited search
def solve_using_iddfs(start_state,goal_state):
    global visited_states_iddfs, timer_iddfs, comparisons_iddfs
    visited_states_iddfs = []
    timer_iddfs = time.time()
    comparisons_iddfs = 0
    for i in range(iddfs_max_depth):
        if depth_limited_dfs(start_state,goal_state,i):
            return True
    timer_iddfs = (time.time() - timer_iddfs) * 10**3
    return False




#Functions to solve using A* Algorithm
visited_states_astar = []


#Fringe to store the possible next states in A*
fringe = pd.DataFrame(columns = ["state","g_value","h_value","f_value"])
fringe["g_value"] = fringe["g_value"].astype('int64')
fringe["h_value"] = fringe["h_value"].astype('int64')
fringe["f_value"] = fringe["f_value"].astype('int64')


#output will store the final steps taken to reach the goal state
output_astar = fringe.copy()

#Function to find the h(n) value of A*
#Heuristic function is taken as number of misplaced tiles
def find_heuristic(state, goal_state):
    array1 = np.array(state)
    array2 = np.array(goal_state)
    
    subtracted_array = np.subtract(array1, array2)
    subtracted = list(subtracted_array)
    
    return int(len(subtracted) - subtracted.count(0))

#g(n) will be the depth of the solution since each step cost will be 1

def solve_using_astar(start_state,goal_state):
    global fringe, output_astar, visited_states_astar, timer_astar
    visited_states_astar = []
    timer_astar = time.time()
    
    h_start = find_heuristic(start_state,goal_state)
    
    fringe.loc[0] = [start_state, 0, h_start, h_start]
    
    while not fringe.empty:
        min_element_index = fringe['f_value'].idxmin()
        min_state = fringe.at[min_element_index,"state"]
        g_value = fringe.at[min_element_index, "g_value"]
        
        entry = fringe.loc[min_element_index]
        output_astar=output_astar.append( [entry] )
        
        #removing the ro from the fringe
        fringe = fringe.drop([min_element_index]).reset_index(drop=True)
    
        #check if this minimum element is the goal state?
        if min_state == goal_state:
            visited_states_astar.append(min_state)
            timer_astar = (time.time() - timer_astar) * 10**3
            print("A Star Goal State Found")
            #Update the UI Grid
            update_grid_astar()
            return True
        
        elif (min_state not in visited_states_astar):
            visited_states_astar.append(min_state)
            neighbouring_states = find_next_moves(min_state)

            for item in neighbouring_states:
                h_value = find_heuristic(item, goal_state)
                fringe.loc[len(fringe.index)] = [item, int(g_value+1), h_value, int(g_value + 1)+h_value]
    
    return False




#Function to solve using IDA*
id_threshold = 0
max_id_threshold = 50

#Fringe to store the possible next states in A*
fringe_id = pd.DataFrame(columns = ["state","g_value","h_value","f_value"])
fringe_id["g_value"] = fringe_id["g_value"].astype('int64')
fringe_id["h_value"] = fringe_id["h_value"].astype('int64')
fringe_id["f_value"] = fringe_id["f_value"].astype('int64')

#output will store the final steps taken to reach the goal state
output_idastar = fringe_id.copy()

#Underlying logic for running IDA* - that is A* with a max f(n) limit
def idastar_algo(start_state,goal_state):
    global fringe_id, output_idastar, visited_states_idastar, visited_states_idastar_local,timer_idastar,id_threshold
    visited_states_idastar_local = []
    
    #Emptying the fringe for every new iteration of IDA*
    fringe_id = fringe_id[0:0]
    
    h_start = find_heuristic(start_state,goal_state)
    
    fringe_id.loc[0] = [start_state, 0, h_start, h_start]
    
    while not fringe_id.empty:
        min_element_index = fringe_id['f_value'].idxmin()
        min_state = fringe_id.at[min_element_index,"state"]
        g_value = fringe_id.at[min_element_index, "g_value"]
        min_f_value = fringe_id.at[min_element_index, "f_value"]
        
        #Condition to check if current min value is greater than allowed threshold or not
        if (min_f_value > id_threshold):
            return min_f_value
        
        entry = fringe_id.loc[min_element_index]
        output_idastar=output_idastar.append( [entry] )

        #removing the ro from the fringe
        fringe_id = fringe_id.drop([min_element_index]).reset_index(drop=True)
    
        #check if this minimum element is the goal state?
        if min_state == goal_state:
            visited_states_idastar.append(min_state)
            return 111111
        
        elif (min_state not in visited_states_idastar_local):
            visited_states_idastar.append(min_state)
            visited_states_idastar_local.append(min_state)
            neighbouring_states = find_next_moves(min_state)

            for item in neighbouring_states:
                h_value = find_heuristic(item, goal_state)
                fringe_id.loc[len(fringe_id.index)] = [item, int(g_value+1), h_value, int(g_value + 1)+h_value]
    
    return min_f_value



whileer = 0

#Driver Function for Iterative Deepening A*
def solve_using_idastar(start_state,goal_state):
    global id_threshold, visited_states_idastar, timer_idastar

    id_threshold = find_heuristic(start_state,goal_state)
    visited_states_idastar = []
    timer_idastar = time.time()
    
    while(id_threshold <= max_id_threshold):
        updated_threshold = idastar_algo(start_state,goal_state)
        
        if updated_threshold == 111111:
            timer_idastar = (time.time() - timer_idastar) * 10**3
            print("IDA* Star Goal State Found")
            #Update the UI Grid
            update_grid_idastar()
            return True
        
        id_threshold = updated_threshold

        
    return False
    


#Function for Branch and Bound
#Why is it similar to A* ? --> Branch and Bound looks for the lower bound of cost which is of the style:
#
# c = g(n) + h(n)
#Here, since each step cost is 1 and the heuristic function is same, i.e number of misplaced tiles in both A* and Branch and Bound, 
#The overall algorithm and performance is also similar

#Fringe to store the possible next states in Branch and Bound
fringe_bb = pd.DataFrame(columns = ["state","g_value","h_value","f_value"])
fringe_bb["g_value"] = fringe_bb["g_value"].astype('int64')
fringe_bb["h_value"] = fringe_bb["h_value"].astype('int64')
fringe_bb["f_value"] = fringe_bb["f_value"].astype('int64')

#output will store the final steps taken to reach the goal state
output_bb = fringe_bb.copy()

def solve_using_branch_bound(start_state,goal_state):
    global fringe_bb, output_bb, visited_states_bb, timer_bb
    visited_states_bb = []
    timer_bb = time.time()
    
    h_start = find_heuristic(start_state,goal_state)
    
    fringe_bb.loc[0] = [start_state, 0, h_start, h_start]
    
    while not fringe.empty:
        min_element_index = fringe_bb['f_value'].idxmin()
        min_state = fringe_bb.at[min_element_index,"state"]
        g_value = fringe_bb.at[min_element_index, "g_value"]
        
        entry = fringe_bb.loc[min_element_index]
        output_bb=output_bb.append( [entry] )
        
        #removing the ro from the fringe
        fringe_bb = fringe_bb.drop([min_element_index]).reset_index(drop=True)

    
        #check if this minimum element is the goal state?
        if min_state == goal_state:
            visited_states_bb.append(min_state)
            timer_bb = (time.time() - timer_bb) * 10**3
            print("A Star Goal State Found")
            #Update the UI Grid
            update_grid_bb()
            return True
        
        elif (min_state not in visited_states_bb):
            visited_states_bb.append(min_state)
            neighbouring_states = find_next_moves(min_state)
            for item in neighbouring_states:
                h_value = find_heuristic(item, goal_state)
                fringe_bb.loc[len(fringe_bb.index)] = [item, int(g_value+1), h_value, int(g_value + 1)+h_value]
    
    return False


#Function to actually apply algorithms
def button_solve(b1,b2,b3,b4,b5,b6,b7,b8,b9):
    if check_input_tiles():
        goal_state = [1,2,3,4,5,6,7,8,9]
        start_state = convert_user_input_to_start_state(b1,b2,b3,b4,b5,b6,b7,b8,b9)

        
        #Disabling the tiles
        global button_variables
        for item in button_variables:
            globals()[item]["state"] = "disabled"
            
            
        #A STAR IMPLEMENTATION
        astar_done = solve_using_astar(start_state,goal_state)
        global visited_states_astar, timer_astar
        if astar_done:
            print("A* Traversal successful.")
            astar_text_comparison = f"""-----A* -----\n Yes\n O(b^d)\n O(b^d)\n {round(timer_astar,3)}ms\n Yes\n {len(visited_states_astar)}\n Yes"""
        
        else:
            print("A* Traversal unsuccesful.")
            astar_text_comparison = f"""-----A*-----\n No\n O(b^d)\n O(b^d)\n {round(timer_astar,3)}ms\n Yes\n {len(visited_states_astar)}\n Yes"""
         
        #root.update_idletasks()
        globals()["label_comparison"]["text"] = "=====SUMMARY & COMPARISON====="
        globals()["label_comparison_targets"]["text"] = "\n-Solution Found:\n-Time Complexity:\n-Space Complexity:\n-Total Time taken to find solution:\n-Complete:\n-Total States Visited:\n-Optimal:"
        globals()["label_comparison2"]["text"] = astar_text_comparison
        root.update()
        
        
        #ITERATIVE DEEPENING A STAR IMPLEMENTATION
        idastar_done = solve_using_idastar(start_state,goal_state)
        if idastar_done:
            print("IDA* Traversal successful.")
            idastar_text_comparison = f"""-----IDA*-----\n Yes\n O(b^d)\n O(bd)\n {round(timer_idastar,3)}ms\n Yes\n {len(visited_states_idastar)}\n Yes"""
        
        else:
            print("IDA* Traversal unsuccesful.")
            idastar_text_comparison = f"""-----IDA*-----\n No\n O(b^d)\n O(bd)\n {round(timer_idastar,3)}ms\n Yes\n {len(visited_states_idastar)}\n Yes"""
        
            
        #root.update_idletasks()
        globals()["label_comparison3"]["text"] = idastar_text_comparison
        root.update()
        

        
        #BRANCH AND BOUND IMPLEMENTATION
        bb_done = solve_using_branch_bound(start_state,goal_state)
        
        if bb_done:
            print("Branch and Bound Traversal succesful.")
            bb_text_comparison = f"""-----B&B-----\n Yes\n O(b^d)\n O(b^d)\n {round(timer_bb,3)}ms\n Yes\n {len(visited_states_bb)}\n Yes"""
        
        else:
            print("Branch and Bound Traversal unsuccesful.")
            bb_text_comparison = f"""-----B&B-----\n No\n O(b^d)\n O(b^d)\n {round(timer_bb,3)}ms\n Yes\n {len(visited_states_bb)}\n Yes"""
        
        globals()["label_comparison4"]["text"] = bb_text_comparison
        root.update()
        
        
#Function to quit the Tkinter GUI
def quit_grid():
    root.destroy()
        

#Function to Reset the Tkinter GUI
def reset_grid():
    
    #Reseting the gobal counter variable to 1
    global global_counter
    global_counter = 1
    
    
    #Reseting summary text
    globals()["label_comparison"]["text"] = "Check back here for the Summary and Comparison after clicking on the button below"
    globals()["label_comparison_targets"]["text"] = " "
    globals()["label_comparison2"]["text"] = " "
    globals()["label_comparison3"]["text"] = " "
    globals()["label_comparison4"]["text"] = " "
    
    #Reseting the values to empty of the input grid
    #Reseting the colour of the input grid
    #Enabling the tiles of input grid
    global button_variables
    global button_variables2
    
    for item in button_variables:
        globals()[item]["text"] = " "
        globals()[item].config(bg="Silver")
        globals()[item]["state"] = "normal"
    
    for item in button_variables2:
        globals()[item]["text"] = " "
        globals()[item].config(bg="Silver")
    
        
        
        
#Building different frames for proper alignment
frame_title = LabelFrame(root)
frame_input = LabelFrame(root, padx=50, pady=50)
frame_output_bfs = LabelFrame(root, padx=50, pady=50)
frame_output_dfs = LabelFrame(root, padx=50, pady=50)
frame_output_iddfs = LabelFrame(root, padx=50, pady=50)
frame_output_comparison = LabelFrame(root)

#Assigning frames to proper grid positions
frame_title.grid(row=0, columnspan=4, sticky = tk.W+tk.E)
frame_input.grid(row=1, column=0)
frame_output_bfs.grid(row=1, column=1)
frame_output_dfs.grid(row=1, column=2)
frame_output_iddfs.grid(row=1,column=3)
frame_output_comparison.grid(row=2, columnspan=4, sticky = tk.W+tk.E)


#Label for the title Frame

label_title = Label(frame_title, text= """SHIKHAR ASTHANA'S 8 PUZZLE SOLVER""", font=("Helvetica", 30))
label_title.pack()



#Label for the first frame
label_input = Label(frame_input, text= "Starting State", font=("Helvetica",20))
label_input2 = Label(frame_input, text= "Assign values by clicking tiles.", font=("Helvetica",10))
label_input.grid(row = 0, columnspan=3, sticky = tk.W+tk.E)
label_input2.grid(row = 1, columnspan=3, sticky = tk.W+tk.E)

#Build the grid for 8-Puzzle Input
b1 = Button(frame_input, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", command=lambda: button_click(b1))
b2 = Button(frame_input, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", command=lambda: button_click(b2))
b3 = Button(frame_input, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", command=lambda: button_click(b3))

b4 = Button(frame_input, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", command=lambda: button_click(b4))
b5 = Button(frame_input, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", command=lambda: button_click(b5))
b6 = Button(frame_input, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", command=lambda: button_click(b6))

b7 = Button(frame_input, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", command=lambda: button_click(b7))
b8 = Button(frame_input, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", command=lambda: button_click(b8))
b9 = Button(frame_input, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", command=lambda: button_click(b9))

#Assigning the buttons to the proper grid position
b1.grid(row=2, column=0)
b2.grid(row=2, column=1)
b3.grid(row=2, column=2)

b4.grid(row=3, column=0)
b5.grid(row=3, column=1)
b6.grid(row=3, column=2)

b7.grid(row=4, column=0)
b8.grid(row=4, column=1)
b9.grid(row=4, column=2)




#Label for the BFS frame
label_bfs = Label(frame_output_bfs, text= "8 Puzzle Solution", font=("Helvetica",20))
label_bfs2 = Label(frame_output_bfs, text= "using A* Algorithm", font=("Helvetica",10))
label_bfs.grid(row = 0, columnspan=3, sticky = tk.W+tk.E)
label_bfs2.grid(row = 1, columnspan=3, sticky = tk.W+tk.E)

#Build the grid for 8-Puzzle Output
b11 = Button(frame_output_bfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")
b12 = Button(frame_output_bfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")
b13 = Button(frame_output_bfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")

b14 = Button(frame_output_bfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")
b15 = Button(frame_output_bfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")
b16 = Button(frame_output_bfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")

b17 = Button(frame_output_bfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")
b18 = Button(frame_output_bfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")
b19 = Button(frame_output_bfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")

#Assigning the buttons to the proper grid position
b11.grid(row=2, column=0)
b12.grid(row=2, column=1)
b13.grid(row=2, column=2)

b14.grid(row=3, column=0)
b15.grid(row=3, column=1)
b16.grid(row=3, column=2)

b17.grid(row=4, column=0)
b18.grid(row=4, column=1)
b19.grid(row=4, column=2)



#Label for the DFS frame
label_dfs = Label(frame_output_dfs, text= "8 Puzzle Solution", font=("Helvetica",20))
label_dfs2 = Label(frame_output_dfs, text= "using IDA*", font=("Helvetica",10))
label_dfs.grid(row = 0, columnspan=3, sticky = tk.W+tk.E)
label_dfs2.grid(row = 1, columnspan=3, sticky = tk.W+tk.E)

#Build the grid for 8-Puzzle Output
b21 = Button(frame_output_dfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")
b22 = Button(frame_output_dfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")
b23 = Button(frame_output_dfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")

b24 = Button(frame_output_dfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")
b25 = Button(frame_output_dfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")
b26 = Button(frame_output_dfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")

b27 = Button(frame_output_dfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")
b28 = Button(frame_output_dfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")
b29 = Button(frame_output_dfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")

#Assigning the buttons to the proper grid position
b21.grid(row=2, column=0)
b22.grid(row=2, column=1)
b23.grid(row=2, column=2)

b24.grid(row=3, column=0)
b25.grid(row=3, column=1)
b26.grid(row=3, column=2)

b27.grid(row=4, column=0)
b28.grid(row=4, column=1)
b29.grid(row=4, column=2)






#Label for the IDDFS frame
label_iddfs = Label(frame_output_iddfs, text= "8 Puzzle Solution", font=("Helvetica",20))
label_iddfs2 = Label(frame_output_iddfs, text= "using Branch and Bound", font=("Helvetica",10))
label_iddfs.grid(row = 0, columnspan=3, sticky = tk.W+tk.E)
label_iddfs2.grid(row = 1, columnspan=3, sticky = tk.W+tk.E)

#Build the grid for 8-Puzzle Output
b31 = Button(frame_output_iddfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")
b32 = Button(frame_output_iddfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")
b33 = Button(frame_output_iddfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")

b34 = Button(frame_output_iddfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")
b35 = Button(frame_output_iddfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")
b36 = Button(frame_output_iddfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")

b37 = Button(frame_output_iddfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")
b38 = Button(frame_output_iddfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")
b39 = Button(frame_output_iddfs, text= " ",  font=("Helvetica",20), height = 2, width = 4, bg = "Silver", state = "disabled")

#Assigning the buttons to the proper grid position
b31.grid(row=2, column=0)
b32.grid(row=2, column=1)
b33.grid(row=2, column=2)

b34.grid(row=3, column=0)
b35.grid(row=3, column=1)
b36.grid(row=3, column=2)

b37.grid(row=4, column=0)
b38.grid(row=4, column=1)
b39.grid(row=4, column=2)







#Label for comparison
label_comparison = Label(frame_output_comparison, text="Check back here for the Summary and Comparison after clicking on the button below", font=("Helvetica",15), justify = "center")
label_comparison_targets = Label(frame_output_comparison, text=" ", font=("Helvetica",15), justify = "left")
label_comparison2 = Label(frame_output_comparison, text=" ", font=("Helvetica",15))
label_comparison3 = Label(frame_output_comparison, text=" ", font=("Helvetica",15))
label_comparison4 = Label(frame_output_comparison, text=" ", font=("Helvetica",15))

label_comparison.grid(row=0, columnspan = 4, sticky = tk.E + tk.W)
label_comparison_targets.grid(row=1, column = 0,sticky = tk.E + tk.W)
label_comparison2.grid(row=1, column = 1,sticky = tk.E + tk.W)
label_comparison3.grid(row=1, column = 2,sticky = tk.E + tk.W)
label_comparison4.grid(row=1, column = 3,sticky = tk.E + tk.W)

#A button to start the solution algorithm
b_solve = Button(root, text = "Start Solving!", font=("Helvetica", 20), height = 3, width = 6, bg = "Silver", command = lambda: button_solve(b1["text"],b2["text"],b3["text"],b4["text"],b5["text"],b6["text"],b7["text"],b8["text"],b9["text"])) 
b_solve.grid(row = 3, column=1, columnspan=2, sticky=tk.N+tk.S+tk.E+tk.W)


#A Button to reset the whole UI
b_reset = Button(root, text = "Reset", font=("Helvetica", 20), height = 3, width = 6, bg = "Silver", command = lambda: reset_grid()) 
b_reset.grid(row = 3, column=0, sticky=tk.N+tk.S+tk.E+tk.W)

#A Button to quit
b_quit = Button(root, text = "Quit", font=("Helvetica", 20), height = 3, width = 6, bg = "Silver", command = lambda: quit_grid()) 
b_quit.grid(row = 3, column=3, sticky=tk.N+tk.S+tk.E+tk.W)


root.mainloop()
```
